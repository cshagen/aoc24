```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx

#nodes= [key for key,m in modules.items()]
G=nx.DiGraph()
```

```python
filename="./data/d24-input.txt"
lines =[line.split(' ') for line in open(filename).read().strip().split('\n')]
#display(lines)
section2= False
gates = []
for line,i in lines:
    if line[0] == '':
        section2= True
    elif section2:
        gates.append((line[1],line[0],line[2],line[4]))
        G.add_node(line[0])
        G.add_node(line[2])
        G.add_node(line[4])
        G.add_edge(line[0],line[4])
        G.add_edge(line[2],line[4])
        

```

```python

#subax1 = plt.subplot()
plt.figure(3,figsize=(12,12)) 

nx.draw_spring(G, with_labels=True, font_weight='regular')
```
